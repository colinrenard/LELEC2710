```python
import kwant
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Fundamental constants

pi = np.pi
h = 6.626069e-34    # Planck constant (Js)
e = 1.602e-19      # electron charge (C)
c = 299792458       # Light celerity (m/s)
kB = 1.3806503e-23  # Boltzmann constant (kg m2/Ks2)
m_e = 9.1e-31       # electron mass (kg)
vf = 1e6            # Fermi velocity (m/s)
hbar = h/(2*pi)     # Reduced planck constant

epsilon_0 = 8.85e-12  # Vacuum permittivity (F/m)
epsilon_hBN = 3
epsilon_Ox = 3.8

def toeV(x):
    ### Transform x in joules to electronVolts
    return x/e
```

```python
db = 20e-9 # bottom hBN layer thickness (m)
dOx = 285e-9 # substrate thickness (m)

Cbg = epsilon_0 / (db/epsilon_hBN + dOx/epsilon_Ox)

# We do not have a precise function for the capacitance linked with the CNT, so we use the approximation they provide
r_cnt = 1e-9 # CNT radius (m)
dt = 4e-9 # top hBN thickness (m)
ht = dt + r_cnt
at = np.sqrt(ht**2 - r_cnt**2)
kappa = (ht+at)**2/r_cnt**2

# Here, epsilon_r is set to have a curve resembling the simulation in the article (Ccnt/e evaluated at 0 is around 2.5e12 cm^-2 V^-1)
def computeCcnt(y, epsilon_r=2.5):
    return epsilon_0*epsilon_r*4*at/((y**2+at**2)*np.log(kappa)) # Try log10 if it does not work

# Modify this function to have the potential as a function of Vbg and Vcnt
def potential(site, Vcnt, Vbg):
    (x, y) = site.pos
    yeff = y*1e-9/(4*np.sqrt(3))
    Ccnt = computeCcnt(yeff)
    n = (Cbg*Vbg + Ccnt*Vcnt)/e
    U = -np.sign(n)*hbar*vf*np.sqrt(pi*np.abs(n))
    # print("Ccnt/e: {}, U: {}, n: {}".format(Ccnt/e, toeV(U), n))
    return toeV(U)
```

```python
# Definition of the system, with a scattering region and some leads
# We work in nanometers and electronvolts

a = 1 # easy values, but should be modified as in the article
t = 1

syst = kwant.Builder()

lattice = kwant.lattice.honeycomb(a) # Graphene lattice (in the practical sessions, norbs=1, why?)
subl1, subl2 = lattice.sublattices # Both sublattices of the graphene lattice

# Definition of the scattering region, square with sides 360 nm. Left and right leads are 34 nm wide
W, L = 36, 36
W_lead = 4

def rectangle(pos, L=L, W=W):
    x, y = pos
    return abs(x) < L//2 and abs(y) < W//2
    
def lead_shape(pos):
    x, y = pos
    return -W_lead//2 < y < W_lead//2

syst[lattice.shape(rectangle, (0,0))] = potential # This is where we can assign a potential depending on both front and backgate voltages. 
# On-site energy without potential is 0 in the case of graphene

hoppings = (((0, 0), subl1, subl2), ((0, 1), subl1, subl2), ((-1, 1), subl1, subl2))
syst[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -t

# Definition of the leads
symLeft = kwant.TranslationalSymmetry(lattice.vec((-1, 0))) # Symmetry of each lead
symRight = kwant.TranslationalSymmetry(lattice.vec((1, 0)))

leadLeft = kwant.Builder(symLeft)
leadLeft[lattice.shape(lead_shape, (0, 0))] = potential # or 0?
leadLeft[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -t

leadRight = kwant.Builder(symRight)
leadRight[lattice.shape(lead_shape, (0, 0))] = potential # or 0?
leadRight[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -t

for lead in [leadLeft, leadRight]:
    syst.attach_lead(lead)
```

```python
# To highlight the two sublattices of graphene, we plot one with
# a filled, and the other one with an open circle:
def family_colors(site):
    return 0 if site.family == subl1 else 1

# Plot the closed system without leads.
kwant.plot(syst, site_color=family_colors, site_lw=0.1, colorbar=False);

syst = syst.finalized()
```

```python
# Computation of the conductance
VcntArray = np.arange(-5, 10, 1)

smatrix = kwant.smatrix(syst, 0, params=dict(Vcnt=5, Vbg=0))
print(smatrix.transmission(1,0))
```
